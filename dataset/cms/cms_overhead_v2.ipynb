```python
import matplotlib.pyplot as plt

plt.style.use('default')
fig = plt.figure(figsize=(4, 3.5))

labels = [
    'δ = 0.1', 'δ = 0.01', 'δ = 0.001', 'δ = 0.0001',
    'δ = 0.1', 'δ = 0.01', 'δ = 0.001', 'δ = 0.0001'
]

comparison = [
    0.008991, 0.013136, 0.017995, 0.024648,
    0.008959, 0.013170, 0.018130, 0.024345
]

# Convert to milliseconds
comparison = [x * 1000 for x in comparison]

colors = [
    '#2ca02c60', '#2ca02c80', '#2ca02ca0', '#2ca02cc0',
    '#ff7f0e60', '#ff7f0e80', '#ff7f0ea0', '#ff7f0ec0'
]

plt.rcParams['font.size'] = 10

bars = plt.bar(range(len(labels)), comparison, color=colors)

plt.ylabel('Average Time (mSec)', fontsize=12)
plt.xlabel('Confidence Level (δ)', fontsize=12)
plt.ylim(0, max(comparison) + 10)
plt.yticks([0, 10, 20, 30], fontsize=11.5)
plt.xticks(range(len(labels)), labels, rotation=45, ha='right', fontsize=11.5)

# Adding legend
handles = [
    plt.Line2D([0], [0], color='#2ca02c', lw=4, alpha=0.6, label='ε = 0.01'),
    plt.Line2D([0], [0], color='#ff7f0e', lw=4, alpha=0.6, label='ε = 0.05')
]
plt.legend(handles=handles, fontsize=11)

# Make inside border (axes spines) thicker
ax = plt.gca()
for spine in ax.spines.values():
    spine.set_linewidth(1.25)  # Set the border width

# Add grid lines for the y-axis
ax.yaxis.grid(True)
ax.xaxis.grid(False)

plt.tight_layout()
plt.savefig('cms_overhead_v2.pdf', dpi=1000)
plt.show()
```

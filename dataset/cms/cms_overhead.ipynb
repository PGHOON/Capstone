```python
import matplotlib.pyplot as plt

plt.style.use('default')
plt.figure(figsize=(7, 3.5))

labels = [
    'δ = 0.1', 'δ = 0.01', 'δ = 0.001', 'δ = 0.0001',
    'δ = 0.1', 'δ = 0.01', 'δ = 0.001', 'δ = 0.0001',
    'δ = 0.1', 'δ = 0.01', 'δ = 0.001', 'δ = 0.0001'
]
#ε: 0.001 --> 0.01 --> 0.05

comparison = [
    0.009067, 0.013360, 0.018085, 0.024978,
    0.008991, 0.013136, 0.017995, 0.024648,
    0.008959, 0.013170, 0.018130, 0.024345
]

colors = [
    '#1f77b460', '#1f77b480', '#1f77b4a0', '#1f77b4c0',
    '#2ca02c60', '#2ca02c80', '#2ca02ca0', '#2ca02cc0',
    '#ff7f0e60', '#ff7f0e80', '#ff7f0ea0', '#ff7f0ec0'
]

plt.rcParams['font.size'] = 10

plt.bar(range(len(labels)), comparison, color=colors)

plt.ylabel('Average Time Spent (Sec)')
plt.ylim(0, max(comparison) + 0.01)
plt.xticks(range(len(labels)), labels, rotation=45, ha='right')

group_positions = [1.5, 5.5, 9.5]
group_labels = ['ε = 0.001', 'ε = 0.01', 'ε = 0.05']

for pos, label in zip(group_positions, group_labels):
    plt.text(pos, max(comparison) + 0.005, label, ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.savefig('cms_overhead.pdf', dpi=1000)
plt.show()


```

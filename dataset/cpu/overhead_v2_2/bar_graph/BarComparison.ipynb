```python
import matplotlib.pyplot as plt
import pandas as pd

baseline_df = pd.read_csv('baseline.csv')
cryptoguard_df = pd.read_csv('cryptoguard.csv')

avg_baseline_cpu_usage = baseline_df['CPU Usage'].mean()
avg_cryptoguard_cpu_usage = cryptoguard_df['CPU Usage'].mean()

labels = ['Baseline', 'CryptoGuard']
average_cpu_usages = [avg_baseline_cpu_usage, avg_cryptoguard_cpu_usage]

plt.rcParams['font.size'] = 60
plt.figure(figsize=(32, 16))
plt.bar(labels, average_cpu_usages, color=['#FFF282', '#45C4B0'], edgecolor='black', lw=4)
plt.ylabel('Average CPU Usage (%)')
plt.ylim(0, max(average_cpu_usages) + 1)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

plt.style.use('default')
# plt.figure(figsize=(5, 4))  # Figure size increased
plt.figure(figsize=(4, 3.5))

baseline = pd.read_csv('baseline.csv')
cryptoguard = pd.read_csv('cryptoguard.csv')

baseline = baseline['CPU Usage']
monitoring = cryptoguard['CPU Usage'][10:45]
detection = cryptoguard['CPU Usage'][45:48]
prevention = cryptoguard['CPU Usage'][48:54]

labels = ['Baseline', 'Monitoring', 'Detection', 'Prevention']
comparison = [baseline, monitoring, detection, prevention]

# Set global font size
ax = plt.gca()
for spine in ax.spines.values():
    spine.set_linewidth(1.25)  # Set the border width

# Add grid lines for the y-axis
ax.yaxis.grid(True)
ax.xaxis.grid(False)

plt.rcParams.update({'font.size': 14})

plt.boxplot(comparison, labels=labels, patch_artist=True,
            boxprops=dict(facecolor='lightgrey', color='black', linewidth=1),
            medianprops=dict(color='red', linewidth=1),
            flierprops=dict(marker='', color='black', alpha=0.25))

plt.ylabel('Average CPU Usage (%)', fontsize=12)
plt.xlabel('Execution Phase', fontsize=12)
plt.xticks(fontsize=11.5, rotation=15)
plt.yticks(fontsize=11.5)
plt.tight_layout()
plt.savefig('average_cpu_comparison_v6.pdf', dpi=1000)
plt.show()
```

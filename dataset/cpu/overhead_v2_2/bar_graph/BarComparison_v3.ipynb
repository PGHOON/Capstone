```python
import matplotlib.pyplot as plt
import pandas as pd

plt.style.use('default')
plt.figure(figsize=(5, 2.5))

baseline = pd.read_csv('baseline.csv')
cryptoguard = pd.read_csv('cryptoguard.csv')

baseline_avg = baseline['CPU Usage'].mean()
cryptoguard_avg = cryptoguard['CPU Usage'][10:54].mean()
detection_avg = cryptoguard['CPU Usage'][10:45].mean()
prevention_avg = cryptoguard['CPU Usage'][45:54].mean()

labels = ['Baseline', 'System call\nMonitoring', 'Detection/\nPrevention', 'CryptoGuard']
comparison = [baseline_avg, detection_avg, prevention_avg, cryptoguard_avg]

print(comparison)

plt.rcParams['font.size'] = 10

plt.bar(labels, comparison, color=['tab:red', 'tab:green', 'tab:olive', 'tab:blue'])

plt.ylabel('Average CPU Usage (%)')
plt.ylim(0, max(comparison) + 1)
plt.tight_layout()
plt.savefig('average_cpu_comparison_v3.pdf', dpi=1000)
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('case2.csv')
test = [df[i:i + 60] for i in range(0, df.shape[0], 60)]

plt.figure(figsize=(32, 16))
for index, chunk in enumerate(test):
    plt.plot(range(1, len(chunk) + 1), chunk['CPU Usage'])

plt.rcParams['font.size'] = 60
plt.xlabel('Time (Second)')
plt.ylabel('CPU Usage (%)')
plt.tight_layout()
plt.show()
```

```python
means = [df['CPU Usage'][i::60].mean() for i in range(60)]

plt.rcParams['font.size'] = 30
plt.figure(figsize=(20, 10))
plt.plot(range(1, 61), means, color='blue')
plt.xlabel('Time')
plt.ylabel('Average CPU Usage')
plt.title('Average CPU Usage for Case (ii)')
plt.xticks(range(0, 61, 1))
plt.yticks(range(0, 25, 5))
plt.tight_layout()
plt.show()
```

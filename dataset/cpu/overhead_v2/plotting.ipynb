```python
import glob
import re
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('case1/case1.csv')
means1 = [df1['cpu usage'][i::60].mean()/10 for i in range(60)]

df2 = pd.read_csv('case2/case2.csv')
means2 = [df2['cpu usage'][i::60].mean()/10 for i in range(60)]

plt.figure(figsize=(20, 10))
plt.plot(range(1, 61), means1, color='red', label='Case (i)')
plt.plot(range(1, 61), means2, color='blue', label='Case (ii)')
plt.legend()
plt.xlabel('Time (Second)')
plt.ylabel('Average CPU Usage %')
plt.title('CPU Usage Over Time')
plt.xticks(range(0, 61, 1))
plt.show()
```

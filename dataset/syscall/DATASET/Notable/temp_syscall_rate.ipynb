# Xmrig Family
23 members --> 230 datasets

```python
import glob
import pandas as pd

csv_files = glob.glob('./data/xmrig/*.csv')

syscall_counts = {}

for file_path in csv_files:
    df = pd.read_csv(file_path)
    for syscall in df['SYSTEM_CALL']:
        if syscall in syscall_counts:
            syscall_counts[syscall] += 1
        else:
            syscall_counts[syscall] = 1

sorted_syscall_counts = dict(sorted(syscall_counts.items(), key=lambda item: item[1], reverse=True))
list(sorted_syscall_counts.items())[:299]
```

# Sysrv Family
100 members --> 984 samples

```python
import glob
import pandas as pd

csv_files = glob.glob('./data/sysrv/*.csv')

syscall_counts = {}

for file_path in csv_files:
    df = pd.read_csv(file_path)
    for syscall in df['SYSTEM_CALL']:
        if syscall in syscall_counts:
            syscall_counts[syscall] += 1
        else:
            syscall_counts[syscall] = 1

sorted_syscall_counts = dict(sorted(syscall_counts.items(), key=lambda item: item[1], reverse=True))
list(sorted_syscall_counts.items())[:299]
```

# Plotting

```python
import matplotlib.pyplot as plt

xmrig_syscalls = [('sys_enter_sched_yield', 3651719),
 ('sys_enter_read', 26613),
 ('sys_enter_futex', 18830),
 ('sys_enter_write', 12388),
 ('sys_enter_epoll_pwait', 8867),
 ('sys_enter_openat', 4048),
 ('sys_enter_close', 3328),
 ('sys_enter_clock_gettime', 2075),
 ('sys_enter_mmap', 1466),
 ('sys_enter_rt_sigprocmask', 1466),
 ('sys_enter_brk', 783),
 ('sys_enter_mprotect', 478),
 ('sys_enter_epoll_ctl', 431),
 ('sys_enter_newfstatat', 405),
 ('sys_enter_writev', 350),
 ('sys_enter_fcntl', 315),
 ('sys_enter_sched_setaffinity', 248),
 ('sys_enter_set_robust_list', 240),
 ('sys_enter_rseq', 240),
 ('sys_enter_clone3', 227),
 ('sys_enter_munmap', 202),
 ('sys_enter_getdents64', 200),
 ('sys_enter_ioctl', 174)]

sysrv_syscalls = [('sys_enter_rt_sigaction', 245263734),
 ('sys_enter_futex', 65807359),
 ('sys_enter_pselect6', 34400042),
 ('sys_enter_epoll_ctl', 30708274),
 ('sys_enter_rt_sigprocmask', 28116221),
 ('sys_enter_close', 22857227),
 ('sys_enter_mmap', 19453332),
 ('sys_enter_sigaltstack', 18104712),
 ('sys_enter_socket', 17259711),
 ('sys_enter_gettid', 16292961),
 ('sys_enter_setsockopt', 16139847),
 ('sys_enter_connect', 13899104),
 ('sys_enter_clone', 9053258),
 ('sys_enter_read', 5568551),
 ('sys_enter_openat', 3744326),
 ('sys_enter_getpid', 3623530),
 ('sys_enter_getppid', 3622755),
 ('sys_enter_bind', 2240276),
 ('sys_enter_munmap', 1924682),
 ('sys_enter_pipe2', 1812125),
 ('sys_enter_execve', 1812059),
 ('sys_enter_chdir', 1811519),
 ('sys_enter_sched_getaffinity', 1811475)]

total_xmrig_calls = sum(value for _, value in xmrig_syscalls)
total_sysrv_calls = sum(value for _, value in sysrv_syscalls)

xmrig_syscalls_percent = [(label, value / total_xmrig_calls * 100) for label, value in xmrig_syscalls]
sysrv_syscalls_percent = [(label, value / total_sysrv_calls * 100) for label, value in sysrv_syscalls]

xmrig_syscalls_sorted = sorted(xmrig_syscalls_percent, key=lambda x: x[1], reverse=True)
sysrv_syscalls_sorted = sorted(sysrv_syscalls_percent, key=lambda x: x[1], reverse=True)

xmrig_labels, xmrig_values = zip(*xmrig_syscalls_sorted)
sysrv_labels, sysrv_values = zip(*sysrv_syscalls_sorted)

colors = plt.cm.tab20.colors
color_dict = {label: colors[i % len(colors)] for i, label in enumerate(set(xmrig_labels) | set(sysrv_labels))}


common_syscalls = set(xmrig_labels) & set(sysrv_labels)
filtered_common_syscalls = {label: min(value for l, value in xmrig_syscalls_percent + sysrv_syscalls_percent if l == label) for label in common_syscalls}

fig, axs = plt.subplots(1, 2, figsize=(16, 8))

#Xmrig
axs[0].pie(xmrig_values, autopct='%1.1f%%', colors=[color_dict[label] for label in xmrig_labels])
axs[0].set_title('XMRig Syscalls')

#Sysrv
axs[1].pie(sysrv_values, autopct='%1.1f%%', colors=[color_dict[label] for label in sysrv_labels])
axs[1].set_title('Sysrv Syscalls')
axs[1].set_position([0.50, 0.1, 0.35, 0.8])

#Legend
all_syscalls_percent = sorted(set(xmrig_syscalls_percent + sysrv_syscalls_percent), key=lambda x: x[1], reverse=True)
sorted_labels = [label for label, value in all_syscalls_percent if label not in filtered_common_syscalls or value > filtered_common_syscalls[label]]
handles = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=color_dict[label], markersize=10) for label in sorted_labels]
fig.legend(handles, sorted_labels, loc='center right', title='Syscall Legend')

for text, label in zip(fig.legends[0].get_texts(), sorted_labels):
    if label in common_syscalls:
        text.set_color('magenta')

plt.savefig('temp_syscall_rate.pdf', dpi=1000, bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

syscall = ['rt_sigaction', 'futex', 'pselect6', 'epoll_ctl', 'rt_sigprocmask', 
    'close', 'mmap', 'sigaltstack', 'socket', 'gettid', 'setsockopt', 'connect',
    'clone', 'read', 'sched_yield', 'openat', 'getpid', 
    'getppid', 'bind', 'munmap', 'execve', 'pipe2', 'sched_getaffinity']
value = [245263906, 65826189, 34400042, 30708705, 28117687, 
    22860555, 19454798, 18104712, 17259858, 16292961, 16139975, 13899219,
    9053396, 5595164, 4641782, 3748374, 3623530, 
    3622755, 2240312, 1924884, 1812059, 1812171, 1811549]


fig, ax = plt.subplots(figsize=(10, 5))
y_pos = np.arange(len(syscall))

ax.barh(y_pos, value, color='gray')
ax.set_yticks(y_pos)
ax.set_yticklabels(syscall, fontsize = 14)
ax.invert_yaxis()
ax.set_xlabel('Count', fontsize = 14)


plt.tight_layout()
plt.savefig('notable_syscalls.pdf', dpi=1000, bbox_inches='tight')
plt.show()
```

```python
NUM_CLASSES = 3
CLASSES = np.array(['benign', 'sysrv', 'xmrig'])
DATASET_DIR = "raw_data/"
VECTOR_LENGTH = 32 * 32

label_encoder = LabelEncoder()
label_encoder.fit(syscalls)

def csvToVector(file_path):
    try:
        data = pd.read_csv(file_path, encoding='utf-8')
    except UnicodeDecodeError:
        data = pd.read_csv(file_path, encoding='ISO-8859-1')
    
    data_encoded = label_encoder.fit_transform(data['SYSTEM_CALL'])
    vector = np.zeros(VECTOR_LENGTH, dtype=np.uint8)
    syscall_nums = min(len(data_encoded), VECTOR_LENGTH)
    vector[:syscall_nums] = data_encoded[:syscall_nums]

    return vector

def process_file(args):
    file_path, class_idx = args
    vector = csvToVector(file_path)
    return vector, class_idx

def load_data(dataset_dir):
    x = []
    y = []
    classes = ["0/60sec_0", "1/60sec_1", "2/60sec_2"]

    file_paths = []
    for class_idx, class_name in enumerate(classes):
        class_dir = os.path.join(dataset_dir, class_name)
        for file_name in os.listdir(class_dir):
            if file_name.endswith('.csv'):
                file_path = os.path.join(class_dir, file_name)
                file_paths.append((file_path, class_idx))

    with Pool() as pool:
        results = pool.map(process_file, file_paths)

    x, y = zip(*results)
    x = np.array(x)
    y = np.array(y)

    return x, y
```

```python
X, y = load_data(DATASET_DIR)
```

```python
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.3, random_state=42)

X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)

X_train = X_train / 299.0
X_val = X_val / 299.0
X_test = X_test / 299.0

y_train = to_categorical(y_train, 3)
y_val = to_categorical(y_val, 3)
y_test = to_categorical(y_test, 3)
```

```python
cp_model = load_model('60_LSTM_checkpoint.h5')
```

```python
y_pred_proba = cp_model.predict(X_test)

y_test_binary = np.argmax(y_test, axis=1)

n_classes = y_test.shape[1]
fpr = dict()
tpr = dict()
roc_auc = dict()

y_test_binarized = label_binarize(y_test_binary, classes=[0, 1, 2])

for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_binarized[:, i], y_pred_proba[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])
    
class_labels = ['Benign', 'Sysrv', 'Xmrig']    
    
plt.figure()
for i in range(n_classes):
    plt.plot(fpr[i], tpr[i], label=f'{class_labels[i]}')

plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc="lower right")
plt.savefig('ROC_60_LSTM.pdf', dpi=1000)
plt.show()

for i in range(n_classes):
    print(f'{class_labels[i]} AUC: {roc_auc[i]:.2f}')
```

```python
# 예측 확률 및 테스트 데이터 준비
y_pred_proba = cp_model.predict(X_test)
y_test_binary = np.argmax(y_test, axis=1)
n_classes = y_test.shape[1]

fpr = dict()
tpr = dict()
roc_auc = dict()

# 테스트 데이터를 이진화
y_test_binarized = label_binarize(y_test_binary, classes=[0, 1, 2])

# ROC 곡선을 계산하고 보간하여 부드럽게 함
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_binarized[:, i], y_pred_proba[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

    # 보간: 1000개의 점으로 보간하여 부드러운 곡선 생성
    all_fpr = np.linspace(0, 1, 1000)
    tpr[i] = interp(all_fpr, fpr[i], tpr[i])
    fpr[i] = all_fpr

class_labels = ['Benign', 'Sysrv', 'Xmrig']

# ROC 곡선 시각화
plt.figure()
for i in range(n_classes):
    plt.plot(fpr[i], tpr[i], label=f'{class_labels[i]} (AUC = {roc_auc[i]:.2f})')

plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc="lower right")
plt.savefig('ROC_60_LSTM.pdf', dpi=1000)
plt.show()

# AUC 값 출력
for i in range(n_classes):
    print(f'{class_labels[i]} AUC: {roc_auc[i]:.2f}')
```
